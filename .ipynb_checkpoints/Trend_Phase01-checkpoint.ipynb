# Define functions

```python
def google_api(data_feed):
    df = pd.DataFrame()
    pytrend = TrendReq(hl='en-US',tz=360 )
    #pytrends = TrendReq(hl='en-US', tz=360, timeout=(10,25), proxies=['https://34.203.233.13:80'], retries=5, backoff_factor=0.1)
    for i in tqdm(range(5,len(data_feed),5)):
        time.sleep(1.5)
        kw = data_feed[i-5:i]
        pytrend.build_payload(kw_list=kw,cat=0,timeframe='2020-01-01 2020-04-01',geo='TW',gprop='')
        data = pytrend.interest_over_time()
        df = pd.concat([df,data],axis=1,sort=False)
        
    df.drop(['isPartial'],axis=1,inplace=True)
    return df 
```

```python
def xgb(x,y):
    model = XGBClassifier()
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    predictions = [round(value) for value in y_pred]
    accuracy = accuracy_score(y_test, predictions)
    print("Accuracy: %.2f%%" % (accuracy * 100.0))
   
```

```python
def gbc(x,y):
    state = 12  
    test_size = 0.2 
  
    X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=test_size, random_state=state)
    lr_list = [0.05, 0.075, 0.1, 0.25, 0.5, 0.75, 1]
    for learning_rate in lr_list:
        gb_clf = GradientBoostingClassifier(n_estimators=20, learning_rate=learning_rate, max_features=2, max_depth=2, random_state=0)
        gb_clf.fit(X_train, y_train)
        print("Learning rate: ", learning_rate)
        print("Accuracy score (training): {0:.3f}".format(gb_clf.score(X_train, y_train)))
        print("Accuracy score (validation): {0:.3f}".format(gb_clf.score(X_val, y_val)))
    
    #filename = "gbc_model.sav"
    #pickle.dump(gb_clf,open(filename,'wb'))
    joblib.dump(gb_clf,filename)
    
```

```python
def call_gbc():
    filename = "gbc_model_1.sav"
    load_model = pickle.load(open(filename,'wb'))
    return load_model
```

```python
def linear_model(x,y):
    X_train, X_test, y_train, y_test = train_test_split(x[['past_Q_average','Trend_now']], y['result'], test_size=0.33, random_state=42)
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    y_pred = lr.predict(X_test)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    
    return y_pred  
```

```python
def feature_engineering(df):
    
    print(df.isnull().sum())
    print(df.shape)
    df.reset_index(level=0,inplace=True)
    df.rename(columns={'date':'Date'},inplace=True)
    df.index = np.arange(1,len(df)+1)
    df['Date'] = df['Date'].astype(str).apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d').date()) 
    df['Year'] = df['Date'].apply(lambda x: datetime.datetime.strftime(x, '%Y')).astype('int64')
    df['Month'] = df['Date'].apply(lambda x: datetime.datetime.strftime(x, '%m')).astype('int64')
    df['Day'] = df['Date'].apply(lambda x: datetime.datetime.strftime(x, '%d')).astype('int64')
    
    return df
    
```

```python
def train_dataframe(df):

    x = df[:66] 
    x.drop(['Date'],axis=1,inplace=True)
    feature = {'Word':[],'past_Q_average':[],'Trend_now':[]}
    prediction_y = []
    columns = x.columns
    for c in columns:
        feature['Word'].append(c)
        feature['past_Q_average'].append(sum(x[c])/len(x))   
    
    trend_now = df[42:66]
    trend_now.drop(['Date'],axis=1,inplace=True)  
    for c in columns:
        num = 0
        for d in trend_now[c]:
            num+=d
        feature['Trend_now'].append(num/len(trend_now))
    feature = pd.DataFrame(feature)
    
    '''
        Setting Prediction -Y 
        this week average trend degree 
    '''
    target_prediction = df[66:]
    target_prediction.drop(['Date'],axis=1,inplace=True)
    for c in columns:
        num = 0
        for d in target_prediction[c]:
            num+=d
        prediction_y.append(num/len(target_prediction))
    #target_prediction = target_prediction.T
    #print(target_prediction)
    y = pd.DataFrame(prediction_y)
    
    
    
    classification = []
    for i,j in zip(prediction_y,feature['Trend_now']):
        if i-j>0:
            
            classification.append(1)
        else:
            classification.append(0)
            
    y['result'] = classification  
    
     
    return feature,y 
```
